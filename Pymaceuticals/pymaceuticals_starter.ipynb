## Bar plots

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas
file = "data/Mouse_metadata.csv"
data = pd.read_csv(file)
df = pd.DataFrame(data)
amounts = df['Drug Regimen'].value_counts().plot(kind='bar',figsize=(14,8),title="Drug Regimens")
amounts.set_xlabel("Drug Regimen")
amounts.set_ylabel("Frequency")
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pyplot

file = "data/Mouse_metadata.csv"
data = pd.read_csv(file)
df = pd.DataFrame(data)
plt.bar(df["Drug Regimen"], df["Drug Regimen"])
plt.xlabel("Drug Regimen")
plt.ylabel("Count")
plt.title("Drug Regimens")
```

## Pie plots

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
file = "data/Mouse_metadata.csv"
data = pd.read_csv(file)
df = pd.DataFrame(data)
plt.title("Male vs Female Mice")
plt.pie('Sex',
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
file = "data/Mouse_metadata.csv"
data = pd.read_csv(file)
df = pd.DataFrame(data)
plt.pie(data,labels=df,autopct='%1.1f%%')
plt.title('My Title')
plt.axis('Sex')
plt.show()
```
